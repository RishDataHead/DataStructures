You are given a binary tree. The binary tree is represented using the TreeNode class. Each TreeNode has an integer value and left and right children, represented using the TreeNode class itself. Convert this binary tree into a binary search tree.

Input:

        10

       /   \

     2      7

   /   \

 8      4

Output:

        8

      /   \

    4     10

  /   \

2      7

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def inorderTraversal(root):
    if root is None:
        return []

    result = []
    stack = []
    current = root

    while True:
        if current is not None:
            stack.append(current)
            current = current.left
        elif stack:
            current = stack.pop()
            result.append(current.value)
            current = current.right
        else:
            break

    return result

def constructBST(root, sorted_list):
    if root is None:
        return

    constructBST(root.left, sorted_list)
    root.value = sorted_list.pop(0)
    constructBST(root.right, sorted_list)

def convertToBST(root):
    if root is None:
        return None

    sorted_list = inorderTraversal(root)
    constructBST(root, sorted_list)

# Create the binary tree from the input
root = TreeNode(10)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(8)
root.left.right = TreeNode(4)

# Convert the binary tree to a binary search tree
convertToBST(root)

# Perform inorder traversal to verify the binary search tree
inorder = inorderTraversal(root)

# Print the inorder traversal result
print("Inorder Traversal of BST:", inorder)
```

ðŸ’¡ Question-3:

Write a program to convert a binary tree to a doubly linked list.

Input:

        10

       /   \

     5     20

           /   \

        30     35

Output:

5 10 30 20 35

</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class DoublyLinkedListNode:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

def convertToDoublyLinkedList(root):
    if root is None:
        return None

    # Create a dummy node as the head of the doubly linked list
    dummy = DoublyLinkedListNode(-1)
    prev = dummy

    def inorderTraversal(node):
        nonlocal prev

        if node is None:
            return

        inorderTraversal(node.left)

        # Update the doubly linked list connections
        doubly_node = DoublyLinkedListNode(node.value)
        prev.next = doubly_node
        doubly_node.prev = prev
        prev = doubly_node

        inorderTraversal(node.right)

    inorderTraversal(root)

    # Set the head of the doubly linked list
    head = dummy.next
    if head is not None:
        head.prev = None

    return head

# Construct the binary tree
root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(20)
root.right.left = TreeNode(30)
root.right.right = TreeNode(35)

# Convert the binary tree to a doubly linked list
head = convertToDoublyLinkedList(root)

# Traverse the doubly linked list and print the values
current = head
while current is not None:
    print(current.value, end=" ")
    current = current.next
```

Write a program to connect nodes at the same level.

Input:

        1

      /   \

    2      3

  /   \   /   \

4     5 6    7

Output:

1 â†’ -1

2 â†’ 3

3 â†’ -1

</aside>

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
        self.next = None

def connectNodesAtSameLevel(root):
    if root is None:
        return None

    # Perform level order traversal using a queue
    queue = [root]

    while queue:
        level_size = len(queue)

        for i in range(level_size):
            node = queue.pop(0)

            # Set the next pointer for the current node
            if i < level_size - 1:
                node.next = queue[0]
            else:
                node.next = None

            # Enqueue the left and right children of the current node
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

    return root

# Construct the binary tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

# Connect nodes at the same level
root = connectNodesAtSameLevel(root)

# Traverse the tree and print the connected nodes
current = root
while current:
    print(current.value, end=" â†’ ")
    if current.next:
        current = current.next
    else:
        print("-1")
        current = current.left
```
