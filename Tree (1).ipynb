Given a binary tree, your task is to find subtree with maximum sum in tree.

Examples:

Input1 : Â  Â  Â  

       1

     / Â  \

   2Â  Â  Â  3

  / \Â  Â  / \

4 Â  5Â  6 Â  7

Output1 : 28

As all the tree elements are positive, the largest subtree sum is equal to sum of all tree elements.



```python
# Node class definition for the binary tree
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

# Recursive function to find the subtree with maximum sum
def maxSubtreeSum(root):
    global maxSum, maxSubtreeRoot
    if root is None:
        return 0

    left_sum = maxSubtreeSum(root.left)
    right_sum = maxSubtreeSum(root.right)

    subtree_sum = root.value + left_sum + right_sum

    if subtree_sum > maxSum:
        maxSum = subtree_sum
        maxSubtreeRoot = root

    return subtree_sum

# Example usage
# Create the binary tree from the example
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

# Initialize the global variables
maxSum = float('-inf')
maxSubtreeRoot = None

# Find the subtree with maximum sum
maxSubtreeSum(root)

# Print the maximum subtree sum
print("Maximum subtree sum:", maxSum)
```

Construct the BST (Binary Search Tree) from its given level order traversal.

Example:

Input: arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output: BST:

            7

         /Â  Â  \

       4 Â  Â  12

     /Â  \ Â  Â  /

    3 Â  6Â  8

   / Â   / Â    \

 1Â    5  Â    10

</aside>

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def constructBST(arr):
    if not arr:
        return None

    root = Node(arr[0])
    queue = [root]
    i = 1

    while i < len(arr):
        current = queue.pop(0)

        if i < len(arr):
            left_val = arr[i]
            if left_val is not None:
                left_child = Node(left_val)
                current.left = left_child
                queue.append(left_child)
            i += 1

        if i < len(arr):
            right_val = arr[i]
            if right_val is not None:
                right_child = Node(right_val)
                current.right = right_child
                queue.append(right_child)
            i += 1

    return root

def inorderTraversal(root):
    if root is None:
        return []

    result = []
    stack = []
    current = root

    while True:
        if current is not None:
            stack.append(current)
            current = current.left
        elif stack:
            current = stack.pop()
            result.append(current.value)
            current = current.right
        else:
            break

    return result

# Example usage
arr = [7, 4, 12, 3, 6, 8, 1, 5, 10]

bst_root = constructBST(arr)

# Perform inorder traversal to verify the BST
inorder = inorderTraversal(bst_root)

# Print the inorder traversal result
print("Inorder Traversal:", inorder)
```

Given an array of size n. The problem is to check whether the given array can represent the level order traversal of a Binary Search Tree or not.

Examples:

Input1 : arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output1 : Yes

For the given arr[], the Binary Search Tree is:

            7

         /Â  Â  \

       4 Â  Â  12

     /Â  \ Â  Â  /

    3 Â  6Â  8

   / Â   / Â    \

 1Â    5  Â    10

Input2 : arr[] = {11, 6, 13, 5, 12, 10}

Output2 : No

The given arr[] does not represent the level order traversal of a BST.

</aside>

```python
def isValidLevelOrder(arr):
    n = len(arr)
    if n == 0:
        return True

    leftChildIndex = 1
    rightChildIndex = 2

    for i in range(1, n):
        if arr[i] < arr[0]:
            if i >= leftChildIndex:
                return False
        else:
            if i >= rightChildIndex:
                return False

        if i == rightChildIndex:
            leftChildIndex *= 2
            rightChildIndex *= 2

    return True

# Example usage
arr1 = [7, 4, 12, 3, 6, 8, 1, 5, 10]
print("Is valid level order traversal?", isValidLevelOrder(arr1))

arr2 = [11, 6, 13, 5, 12, 10]
print("Is valid level order traversal?", isValidLevelOrder(arr2))
```
